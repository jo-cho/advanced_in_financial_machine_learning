CV을 하는 목적은 무엇인가?

간단히 말하자면 머신러닝 알고리즘의 오버피팅을 막는 것이다.

하지만 금융 머신러닝에서 단순한 CV (K-FOLD)는 실패하기 마련이다.

그 이유는 다음과 같다.

1. observations cannot be assumed to be drawn from an IID process
2. the testing set is used multiple times in the process of developing a model, leading to multiple testing and selection bias.

첫 번째 이유부터 보자

시계열 자료는 autocorrelation을 가지기 마련이다.


하지만 k-fold나 shuffle split CV는 IID를 가정하기 때문에 훈련과 테스트 세트의 correlation 문제를 해결할 수 없다.
